```python
import torch
import matplotlib.pyplot as plt

# Create a random 2D PyTorch tensor
tensor_data = torch.rand(10, 10) # Creates a 10x10 tensor with random values between 0 and 1

# Convert the tensor to a NumPy array for visualization
numpy_array = tensor_data.numpy()

# Visualize the tensor
plt.figure(figsize=(6, 6))
plt.imshow(numpy_array, cmap='viridis')
plt.colorbar(label='Tensor Value')
plt.title('Visualization of a PyTorch Tensor')
plt.xlabel('Column Index')
plt.ylabel('Row Index')
plt.show()

print("Tensor created:\n", tensor_data)
print("Tensor shape:", tensor_data.shape)
```
